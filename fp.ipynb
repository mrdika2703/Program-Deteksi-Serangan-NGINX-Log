```python
import re

# Fungsi untuk membaca file log
def read_log(file_path):
    with open(file_path, 'r') as file:
        return file.readlines()

# Fungsi untuk mendeteksi jenis-jenis serangan dalam log
def detect_attacks(log_lines):
    # Inisialisasi jumlah serangan yang ditemukan
    attacks = {
        "File Inclusion Attempt": 0,
        "SQL Injection": 0,
        "Brute Force Attack": 0,
        "normal": 0
    }

    # Daftar untuk menyimpan detail serangan yang terdeteksi
    details = []

    # Memeriksa setiap baris dalam log
    for line in log_lines:
        # Deteksi File Inclusion Attack berdasarkan pola '../' atau './'
        if re.search(r'\.\./|\./', line):
            attacks["File Inclusion Attempt"] += 1
            details.append(f"[File Inclusion Attempt] {line.strip()}")
        # Deteksi SQL Injection berdasarkan karakter tertentu
        elif re.search(r'(\%27)|(\')|(\-\-)|(\%23)|(#)', line):
            attacks["SQL Injection"] += 1
            details.append(f"[SQL Injection] {line.strip()}")
        # Deteksi Brute Force Attack berdasarkan kata kunci seperti login, admin, user, password
        elif re.search(r'login|admin|user|password', line):
            attacks["Brute Force Attack"] += 1
            details.append(f"[Brute Force Attack] {line.strip()}")
        else:
            attacks["normal"] += 1

    return attacks, details

# Fungsi utama untuk menjalankan program
def main():
    # Membaca baris-baris log dari file
    log_lines = read_log('old2.log')
    # Mendeteksi serangan dari log
    attacks, details = detect_attacks(log_lines)
    
    # Menampilkan hasil deteksi serangan
    print("Deteksi Serangan dari Log Nginx:\n")
    for attack_type, count in attacks.items():
        print(f"{attack_type}: {count} kejadian")
    
    # Menampilkan detail log dengan serangan
    print("\nDetail Log dengan Serangan:")
    for detail in details:
        print(detail)

# Menjalankan fungsi utama
if __name__ == "__main__":
    main()
```
